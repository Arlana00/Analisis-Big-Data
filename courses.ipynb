```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV
data = pd.read_csv('courses.csv')  # ubah 'nama_file.csv' dengan nama file dataset yang ingin di olah

# Melihat 5 data teratas dan terbawah dari dataset
print("\nDataset:")
data
```

```python
# Melihat informasi dasar tentang dataset
print("Informasi Dataset:")
print(data.info())
```

```python
# Statistik deskriptif untuk kolom numerik
print("\nStatistik Deskriptif:")
print(data.describe())
```

```python
# Melihat jumlah missing values pada setiap kolom
print("\nJumlah Missing Values pada Setiap Kolom:")
print(data.isnull().sum())
```

```python
#distribusi modul presentation length
plt.figure(figsize=(8, 6))
sns.histplot(data['module_presentation_length'], bins=10, kde=True)
plt.xlabel('Module Presentation Length')
plt.ylabel('Frequency')
plt.title('Distribution of Module Presentation Length')
plt.show()
```

```python
# countplot ini digunakan untuk menghitung seberapa banyak data kode modul.
plt.figure(figsize=(10, 6))
sns.countplot(x='code_presentation', data=data)
plt.xlabel('Kode_presentation')
plt.ylabel('Jumlah')
plt.title('')
plt.show()
```

```python
# scatter plot di gunakan untuk menampilkan hubungan antara variable kode presentasi dan kode modul serta melihat polasebaran / korelasi.

plt.figure(figsize=(8, 6))
sns.scatterplot(x='code_module', y='code_presentation', data=data)
plt.xlabel('Kode_Modul')
plt.ylabel('kode_presentasi')
plt.title('scatter plot')
plt.show()
```
