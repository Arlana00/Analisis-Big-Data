```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV
data = pd.read_csv('vle.csv')  # ubah 'nama_file.csv' dengan nama file dataset yang ingin di olah

# Melihat 5 data teratas dan terbawah dari dataset
print("\nDataset:")
data
```

```python
# Melihat informasi dasar tentang dataset
print("Informasi Dataset:")
print(data.info())
```

```python
# Statistik deskriptif untuk kolom numerik
print("\nStatistik Deskriptif:")
print(data.describe())
```

```python
# Periksa data yang hilang
missing_data = data.isnull().sum()
print(missing_data)
```

```python
# Menghapus baris dengan nilai yang hilang
data = data.dropna()
```

```python
# Mengganti nilai yang hilang dengan rata-rata kolom
data['week_from'].fillna(data['week_from'].mean(), inplace=True)
data['week_to'].fillna(data['week_to'].mean(), inplace=True)
```

```python
# Plot jumlah aktivitas berdasarkan jenis aktivitas
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='activity_type')
plt.title('Jumlah Aktivitas berdasarkan Jenis Aktivitas')
plt.xticks(rotation=45)
plt.show()
```

```python
# menghitung distribusi jenis modul kode 
plt.figure(figsize=(10, 6))
sns.countplot(data=data, x='code_module')
plt.title('Distribusi Jenis Modul Kode')
plt.xticks(rotation=45)
plt.show()
```
