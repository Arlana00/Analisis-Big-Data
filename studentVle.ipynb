```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset dari file CSV
data = pd.read_csv('studentVle.csv')  # ubah 'nama_file.csv' dengan nama file dataset yang ingin di olah
# Melihat informasi dasar tentang dataset
print("Informasi Dataset:")
print(data.info())

# Statistik deskriptif untuk kolom numerik
print("\nStatistik Deskriptif:")
print(data.describe())

# Melihat 5 data teratas dan terbawah dari dataset
print("\nDataset:")
data
```

```python
# Menghitung jumlah nilai yang hilang
missing_values = data.isnull().sum()
print(missing_values)
```

```python
# Visualisasi hubungan antara "sum_click" dan "date"
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="date", y="sum_click")
plt.title("Hubungan antara date dan sum_click")
plt.xlabel("date")
plt.ylabel("sum_click")
plt.show()
```
